```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("../data/csv/results/babbage_200_pairs.csv").set_index('step')
```

```python
plt.figure(figsize=(10,5))
plt.title("GPT-3 Babbage Training Loss During Fine-tuning")
plt.plot(df['training_loss'].to_numpy(),label="Training Loss")

plt.xlabel("Steps")
plt.ylabel("Loss")
plt.legend()
plt.savefig("../visualizations/babbage_loss.png",dpi=300)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.title("GPT-3 Babbage Training Token Accuracy During Fine-tuning \n(Rolling average with window size = 10)")
plt.plot(df['training_token_accuracy'].rolling(10).mean().to_numpy(),label="Training Token Accuracy")

plt.xlabel("Steps")
plt.ylabel("Accuracy")
plt.legend()
plt.savefig("../visualizations/babbage_token_acc.png",dpi=300)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.title("GPT-3 Babbage Training Sequence Accuracy During Fine-tuning \n(Rolling average with window size = 10)")
plt.plot(df['training_sequence_accuracy'].rolling(10).mean().to_numpy(),label="Training Sequence Accuracy")

plt.xlabel("Steps")
plt.ylabel("Accuracy")
plt.legend()
plt.savefig("../visualizations/babbage_sequence_acc.png",dpi=300)
plt.show()
```
